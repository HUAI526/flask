```python
import requests
#環保署環境資料開放平臺分放資料API,酸雨資料
url = "https://data.epa.gov.tw/api/v1/acidr_p_01?offset=0&limit=1000&api_key=b21c95e8-ca23-4874-a503-c53093866b0c"
```

```python
response = requests.get(url)
alldata = response.json()
```

```python
alllist = alldata["records"]
for item in alllist:
    print(item["Site_Name"])
    #print(item["Result_Value"])
    print("=======================")
```

```python
import sqlite3
con = sqlite3.connect('環保署雨量1.db')
```

```python
cur = con.cursor()
cur.execute('''CREATE TABLE 雨量
            (站名 text, 城市 text, 日期 text, 雨量 text)''')
for item in alllist:
    cur.execute(f"INSERT INTO 雨量 VALUES ('{item['Site_Name']}', '{item['County']}','{item['Monitor_Date']}','{item['Result_Value']}')")
```

```python
con.commit()
con.close()
```
